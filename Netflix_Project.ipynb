```python
Q1 = np.percentile(df['release_year'], 25, method='midpoint')
Q3 = np.percentile(df['release_year'], 75, method='midpoint')
IQR = Q3 - Q1
print(IQR)
```

```python
Q1 = df['release_year'].quantile(0.25)
Q3 = df['release_year'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
print(IQR,lower,upper)
```

```python
upper_array = np.where(df['release_year']>=upper)[0]
lower_array = np.where(df['release_year']<=lower)[0]

df.drop(index=upper_array, inplace=True)
df.drop(index=lower_array, inplace=True)
print("New Shape: ", df.shape)
```
